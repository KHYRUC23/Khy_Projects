```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameter P
lam = 1
```

```python
# Generates exponential RV by performing the inverse transform method
def exp_RV(N):

    return -np.log(np.random.rand(N))
```

```python
N_values = [10,100,1000]
```

```python
for N in N_values:
    # 
    exp = exp_RV(N)
    plt.hist(exp, bins=30, density=True, alpha=0.7, label=f'Empirical Histogram (N={N})')

# Generate exact density for comparison
    exact_dens = np.linspace(0, max(exp), 1000)
    plt.plot(exact_dens, np.exp(-exact_dens), color='red', label='Expected Density')
# Plotting hist
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.title(f'(N={N})Empirical Histogram and Expected Density ')
    plt.legend()
    plt.show()
```
